```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#handling missing values
df['Age']=df['Age'].replace(np.NaN,df['Age'].median())
df.isnull().sum()
```

```python
df=df.dropna(subset=['Embarked'])
df.isnull().sum()
```

```python
titanic=['Ticket','PassengerId','Name','Cabin']
df.drop(titanic,axis=1,inplace=True)
df.head()
```

```python
df.describe()
```
